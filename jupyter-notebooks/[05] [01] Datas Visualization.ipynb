```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_data(df, classes, filename):
    if 'sub_class' in df.columns:
        label  = 'sub_class'
    else:
        label  = 'class'
    
    cnt_be = df[df[label]==classes[0]].shape[0]
    cnt_mw = df[df[label]!=classes[0]].shape[0]

    labels = 'Benigno', 'Malicioso'
    sizes = [cnt_be, cnt_mw]
    explode = (0,0.1)

    # Plotting
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
    
    # Saving
    if filename != None:
        fig1.savefig('../Figures/Data/'+filename, dpi=fig1.dpi)
    
def plot_nw_data(df, classes, filename):
    if 'sub_class' in df.columns:
        label  = 'sub_class'
    else:
        label  = 'class'    
    
    cnt_sp = df[df[label]==classes[1]].shape[0]
    cnt_ph = df[df[label]==classes[2]].shape[0]
    cnt_mw = df[df[label]==classes[3]].shape[0]

    labels = 'Spam', 'Phishing', 'Malware'
    sizes = [cnt_sp, cnt_ph, cnt_mw]
    explode = (0,0,0)

    # Plotting
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()    

    # Saving
    if filename != None:
        fig1.savefig('../Figures/Data/'+filename, dpi=fig1.dpi)    
```

```python
# Dataset

ba_df = pd.read_csv('../Dataset/Domains/Clean/domains.csv')
un_df = pd.read_csv('../Dataset/Domains/Clean/domains_raw.csv')

classes = ['benign','spam','phishing','malware']

# plot_data(ba_df, classes, filename=None)
# plot_nw_data(ba_df, classes, filename=None)

plot_data(un_df, classes, filename='original_data_distribution.png')
plot_nw_data(un_df, classes, filename='original_mw_data_distribution.png')
```

```python
# Train

ba_df = pd.read_csv('../Dataset/Train/train.csv')
# un_df = pd.read_csv('../Dataset/Train/un_train.csv')

classes = [0,1,2,3]

plot_data(ba_df, classes, filename='train_data_distribution.png')
plot_nw_data(ba_df, classes, filename='train_mw_data_distribution.png')

# plot_data(un_df, classes, filename='un_tr')
# plot_nw_data(un_df, classes, filename='un_mw_tr')
```
